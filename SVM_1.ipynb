1. What is the mathematical formula for a linear SVM ?

The mathematical formula for a linear Support Vector Machine (SVM) can be represented as follows:
Given a set of training data points $(x_1,y_1),(x_2,y_2),....,(x_n,y_n)$, where $ x_i$ represents the feature vector of the ith data point and $y_i$ represents its corresponding class label $(y_iϵ${-1,1}) for a binary classification problem), the goal of the linear SVM is to find the optimal hyperplane that separates the data into two classes with the maximum margin .

The equation of the hyperplane can be written as:
$$w^Tx+b=0$$
where:
* w is the weight vector perpendicular to the hyperplane.
* b is the bias term.
* x represents the feature vector of a data point.

For a binary classification problem, the decision rule for classifying a new data point x is determined by the sign of $w^Tx+b$ . If $w^Tx+b>0$ ,the point belongs to one class ;$w^Tx+b<0$ ,it belongs to the other class.

2. What is the objective function of a linear SVM ?

The objective function of a linear Support Vector Machine (SVM) is to find the optimal hyperplane that separates the data into two classes with the maximum margin. Mathematically, the objective function can be formulated as follows:

$$Minimize \frac{1}{2}||w||^2$$

Where:
* w is the weight vector perpendicular to the hyperplane.
* ∥w∥ represents the Euclidean norm (magnitude) of the weight vector w.

This objective function aims to minimize the squared Euclidean norm of the weight vector. Minimizing $||w||^2$effectively maximizes the margin of separation between the classes.

In SVM, maximizing the margin corresponds to minimizing $||w||^2$subject to the constraint that all data points are correctly classified and lie on the correct side of the hyperplane with a margin of at least 1. These constraints are formulated as:

$y_i(w^Tx+b)≥1$, for i=1,2,....,n

Where:
* $x_i$ represents the feature vector of the ith data point.
* $y_i$ represents the corresponding class label ($y_iϵ${-1,1}for binary classification).
* b is the bias term.

The objective function and these constraints together define the optimization problem of the linear SVM, which aims to find the optimal hyperplane that separates the classes while maximizing the margin. This optimization problem is typically solved using techniques like Quadratic Programming (QP) or Sequential Minimal Optimization (SMO).

3. What is the kernel trick in SVM ?

he kernel trick is a method used in Support Vector Machines (SVMs) to handle non-linearly separable data by implicitly mapping the input features into a higher-dimensional space without explicitly calculating the transformation. This is achieved by using a kernel function.

In a traditional SVM, the decision boundary (hyperplane) is linear, which means it can only separate data that is linearly separable. However, many real-world datasets are not linearly separable, and trying to fit a linear boundary may lead to poor classification performance.

The kernel trick allows SVMs to implicitly operate in a higher-dimensional feature space where the data might be more easily separable. Instead of explicitly transforming the input features into this higher-dimensional space, the kernel function computes the dot product between the feature vectors in the higher-dimensional space directly in the original feature space.

4. What is the role of support vectors in SVM . Explain with example .

In Support Vector Machines (SVMs), support vectors play a crucial role in defining the decision boundary (hyperplane) and determining the classification of data points. Support vectors are the data points that lie closest to the decision boundary, and they directly influence the position and orientation of the hyperplane.

Let's illustrate the role of support vectors with an example:

Suppose we have a binary classification problem with two classes, represented as red circles and blue crosses in a two-dimensional feature space:

* Red circles: Class A

* Blue crosses: Class B

In the above image, the solid line represents the decision boundary (hyperplane) determined by the SVM. The two dashed lines parallel to the decision boundary represent the margin, which is the distance between the decision boundary and the closest data points from each class. The data points lying on these dashed lines are the support vectors.

In this example:

* The red circle closest to the decision boundary and the blue cross closest to the decision boundary are the support vectors for their respective classes.
* These support vectors define the margin, and the hyperplane is positioned to maximize this margin while separating the classes.
* The decision boundary is determined by the support vectors and is positioned equidistant from them.

During the training process of the SVM, only the support vectors are critical for defining the hyperplane, while the other data points are irrelevant for determining the position of the hyperplane as long as they do not lie on the margin or misclassified.

Support vectors are essential for the generalization ability of SVMs because they represent the critical data points that determine the decision boundary. Even if the dimensionality of the feature space is very high, the number of support vectors is typically small, making SVMs computationally efficient and memory-efficient.







5. Illustrate with examples and graphs of hyperplane , marginal plane , soft margin and hard margin in SVM .

Let's illustrate the concepts of hyperplane, margin, soft margin, and hard margin in SVM with examples and corresponding graphs.

1. Hyperplane:
* The hyperplane is the decision boundary that separates the data points into different classes.
* In a binary classification problem, it is a linear boundary in the feature space.
* Example: Consider a dataset with two classes, represented by red circles and blue crosses.

2. Margin:
* The margin is the region between the hyperplane and the closest data points from each class.
* In a hard-margin SVM, the margin is maximized, and the hyperplane is positioned to have equal distance from the closest data points from each class.
* Example: The dashed lines represent the margin, and the support vectors are the data points lying on these lines.

3. Hard Margin vs. Soft Margin:
* In a hard-margin SVM, the goal is to find a hyperplane that perfectly separates the data with no misclassifications.
* In a soft-margin SVM, some misclassifications are allowed to achieve a better generalization when the data is not perfectly separable.
* Example: In a hard-margin SVM, all data points are correctly classified with a large margin. In a soft-margin SVM, some data points may lie within the margin or even on the wrong side of the margin.

4. Effect of Outliers:
* Outliers can significantly affect the decision boundary in SVM.
* In a hard-margin SVM, outliers can make the model less robust by forcing the hyperplane to be too sensitive to small changes in the data.
* In a soft-margin SVM, the model can handle outliers better by allowing some misclassifications.
* Example: In a hard-margin SVM, an outlier (represented by the green square) forces the hyperplane to change significantly, leading to misclassification of other data points.

These examples illustrate how the concepts of hyperplane, margin, and margin types (hard vs. soft) work in SVM and how they can be affected by outliers. In practice, choosing the appropriate margin type depends on the specific characteristics of the dataset and the desired trade-off between model complexity and generalization ability.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate example data
np.random.seed(0)
X = np.random.randn(20, 2)
y = np.concatenate([np.ones(10), -np.ones(10)])

# Hard Margin SVM
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, s=100)

# Plot the hyperplane
w = np.array([0.5, -0.2])  # example weight vector
b = 0  # example bias term
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = np.dot(np.c_[xx.ravel(), yy.ravel()], w) + b
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
            linestyles=['--', '-', '--'])

plt.title('Hard Margin SVM')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

6. SVM implementation through Iris dataset .

i) Load the iris dataset from sckit-learn library and split into training set and testing set.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

ii) Train a linear SVM classifier on the training set and predict the labels for the testing set .

```python
# Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Initialize SVM classifier
svm_classifier = SVC(kernel='linear', C=1.0, random_state=42)

# Train the SVM classifier
svm_classifier.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = svm_classifier.predict(X_test)
y_pred
```

iii) Compute the accuracy of the model on the testing set .

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

iv) Plot the decision boundaries of the trained model using two of the feature .

```python
# Plotting decision regions
plt.figure(figsize=(10, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('SVM decision regions for Iris dataset')
plt.show()
```

v) Try different value of the regularisation parameter C and see how it affects the pwrformance of model .

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Define a range of values for the regularization parameter C
C_values = [0.001, 0.01, 0.1, 1, 10, 100, 1000]

# Train SVM classifiers with different values of C and evaluate performance
accuracy_scores = []
for C in C_values:
    # Initialize SVM classifier with linear kernel and current C value
    svm_classifier = SVC(kernel='linear', C=C, random_state=42)

    # Train the SVM classifier
    svm_classifier.fit(X_train, y_train)

    # Make predictions on the testing set
    y_pred = svm_classifier.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Plot accuracy scores vs. values of C
plt.figure(figsize=(10, 6))
plt.plot(C_values, accuracy_scores, marker='o', linestyle='-')
plt.xscale('log')
plt.xlabel('Regularization Parameter (C)')
plt.ylabel('Accuracy')
plt.title('Effect of Regularization Parameter C on SVM Performance')
plt.grid(True)
plt.show()
```
